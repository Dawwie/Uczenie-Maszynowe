```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv('data6.tsv', header=None, sep='\t')
```

```python
X = np.array(df[0]).reshape(-1, 1)
y = np.array(df[1])

# Punkt A
```

```python
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.show()
```

```python
poly = PolynomialFeatures(2)
X2 = poly.fit_transform(X)
```

```python
model2 = LinearRegression().fit(X2, y)
y_pred2 = model2.predict(X2)


# Potrzeba posegregować dane, żeby pyplot dobrze narysował wykres

X_reshaped = X.reshape(20,)
sorted_index = np.argsort(X.reshape(20,))
X_reshaped = X_reshaped[sorted_index]
y_pred2 = y_pred2[sorted_index]
```

```python
plt.scatter(X, y)
plt.plot(X_reshaped, y_pred2)
plt.show()
```

```python
poly5 = PolynomialFeatures(5)
X5 = poly5.fit_transform(X)
```

```python
model5 = LinearRegression().fit(X5, y)
y_pred5 = model5.predict(X5)

X_reshaped = X.reshape(20,)
sorted_index = np.argsort(X_reshaped)
X_reshaped = X_reshaped[sorted_index]
y_pred5 = y_pred5[sorted_index]
```

```python
plt.scatter(X, y)
plt.plot(X_reshaped, y_pred5)
plt.show()

# Nie występuje zjawisko nadmiernego dopasowania
```
